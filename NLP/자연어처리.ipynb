```python
from konlpy.tag import Kkma
```

```python
kkma = Kkma()
```

```python
import pandas as pd

data = pd.read_csv("plan_youtube_comment.csv", encoding="utf-8-sig")
```

```python
print(kkma.nouns(data["comment"][0]))
```

```python
nouns_youtube = []
for i in data["comment"] :
  nouns_youtube.extend(kkma.nouns(i))
```

```python
from nltk.tag import pos_tag
from nltk import FreqDist
```

```python
fd_youtube = FreqDist(nouns_youtube)
```

```python
len(fd_youtube)
```

```python
fd_youtube.most_common(30)
```

```python
import numpy as np
```

```python
comment_df = pd.DataFrame(np.c_[[i for i in fd_youtube.most_common()]], columns=["comment", "count"])
```
