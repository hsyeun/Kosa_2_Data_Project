```python
df1 = pd.DataFrame(columns=["횟수", "국가", "출전연도"])
name = list(set(본선진출국가["대륙"]))
count=0
for i in (아시아, 북중미, 아프리카, 남미, 유럽):
    i["대륙"] = name[count]
    df1 = pd.concat([df1, i], ignore_index=True)
    count += 1
```

```python
df1
```

```python
df1["2018 출전"] = 0
for i in range(len(df1)) :
    if df1["국가"][i] in list(본선진출국가["국가명"]) :
        df1["2018 출전"][i] = 1
```
